```python
import altair as alt
import ase.db
import numpy as np
import pandas as pd
```

```python
alt.data_transformers.enable('data_server')
```

```python
db = ase.db.connect('structures.db')
```

```python
# Import full dataset, which is basically the ASE DB in dataframe format
df_full = pd.read_csv('dataset.csv')
# Compute atomic density for each structure
df_full['at_density'] = [db.get(hash=row[1].hash).natoms / db.get(hash=row[1].hash).volume for row in df_full.iterrows()]
df_full
```

```python
df_full_long = pd.melt(df_full.reset_index(), id_vars=['index', 'hash', 'type', 'energy', 'force0', 'force1', 'force2', 'element', 'at_density'])
df_full_long['variable'] = pd.to_numeric(df_full_long['variable']).to_numpy() - 6
df_full_long['descriptor_type'] = [2 if i <= 21 else 3 for i in df_full_long.variable]
df_full_long
```

```python
# DF with testing dataset removed.
df_full_long_no_testing = df_full_long[df_full_long['index'] >= 250]
#df_full_long_no_testing[df_full_long_no_testing['index'] == 250]
```

```python
al_indices = pd.read_csv('train_indices.csv')
```

```python
df_al = df_full.iloc[al_indices.to_numpy().flatten()]
df_al = df_al.reset_index()
df_al['al_iteration'] = [int(np.floor(row[0] / 1000)) for row in df_al.iterrows()]
df_al_long = df_full_long.loc[df_full_long['index'].isin(al_indices.to_numpy().flatten())]
# TODO: add al iteration as a column in dataframe
df_al_long['al_iteration'] = [int(df_al[df_al['index'] == row[1]['index']].al_iteration) for row in df_al_long.iterrows()]
df_al_long
```

```python
# AL CHART
al_chart_by_type = alt.Chart().transform_density(
    'value',
    as_=['value', 'density'],
    #extent=[0, 1.8],
    groupby=['variable', 'type']
).mark_area(
    orient='vertical', 
    opacity=0.3, 
    clip=True
).encode(
    x=alt.X(
        'value:Q',
    ),
    y=alt.Y(
        'density:Q',
        stack='zero',
        impute=None,
        title=None,
        #axis=alt.Axis(labels=False, values=[0],grid=False, ticks=True),
        #scale=alt.Scale(domain=[25,32]),
    ),
    color='type:N',
)

al_chart_tot = alt.Chart().transform_density(
    'value',
    as_=['value', 'density'],
    #extent=[0, 1.8],
    groupby=['variable']
).mark_line(
    orient='vertical',
    clip=True
).encode(
    x=alt.X(
        'value:Q',
        scale=alt.Scale(domain=[0,2])
    ),
    y=alt.Y(
        'density:Q',
        stack='zero',
        impute=None,
        title=None,
        #axis=alt.Axis(labels=False, values=[0],grid=False, ticks=True),
        scale=alt.Scale(domain=[0,6]),
    ),
    color=alt.value('black')
)

# FULL CHART
full_chart_by_type = alt.Chart().transform_density(
    'value',
    as_=['value', 'density'],
    #extent=[0, 2],
    groupby=['variable', 'type']
).mark_area(
    orient='vertical',
    opacity=0.3,
    clip=True
).encode(
    x=alt.X(
        'value:Q',
    ),
    y=alt.Y(
        'density:Q',
        stack='zero',
        impute=None,
        title=None,
        #axis=alt.Axis(labels=False, values=[0],grid=False, ticks=True),
        #scale=alt.Scale(domain=[25,32]),
    ),
    color='type:N',
)

full_chart_tot = alt.Chart().transform_density(
    'value',
    as_=['value', 'density'],
    #extent=[0, 2],
    groupby=['variable']
).mark_line(
    orient='vertical',
    clip=True
).encode(
    x=alt.X(
        'value:Q',
        scale=alt.Scale(domain=[0,2]),
    ),
    y=alt.Y(
        'density:Q',
        stack='zero',
        impute=None,
        title=None,
        #axis=alt.Axis(labels=False, values=[0],grid=False, ticks=True),
        scale=alt.Scale(domain=[0,6]),
    ),
    color=alt.value('black')
)
```

```python
descriptors_to_compare = [1,13,20]
#descriptors_to_compare = [22,26,31]

al_layered = alt.layer(
    al_chart_tot,
    al_chart_by_type,
    #data=df_al_long[df_al_long.variable != 15]
    data=df_al_long[df_al_long.variable.isin(descriptors_to_compare)]
).facet(
    column=alt.Column(
        'variable:O',
        #center=True,
        header=alt.Header(
            titleOrient='bottom',
            labelOrient='bottom',
            labelPadding=0,
        ),
    ),
)

full_layered = alt.layer(
    full_chart_tot,
    full_chart_by_type,
    data=df_full_long[df_full_long.variable.isin(descriptors_to_compare)]
).facet(
    column=alt.Column(
        'variable:O',
        #center=True,
        header=alt.Header(
            titleOrient='bottom',
            labelOrient='bottom',
            labelPadding=0,
        ),
    ),
)
```

```python
alt.vconcat(full_layered, al_layered)
```
