```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def gaussian_II(r, eta, offset, cutoff):
    def cutoff_func(r, cutoff):
        f =  0.5 * (1 + np.cos(np.pi * r / cutoff))
        f[r > cutoff] = 0
        return f

    g = np.exp(-eta * (r - offset)**2 / cutoff**2) * cutoff_func(r, cutoff)

    return g
```

```python
etas = [1, 20]
cutoff = 3
offsets = np.array([0, 0.3, 0.5, 0.7, 0.8]) * cutoff
r_space = np.linspace(0, 3, 100)
```

```python
fig, ax = plt.subplots()

for e in etas:
    for o in offsets:
        ax.plot(r_space, gaussian_II(r_space, e, o, cutoff))
```
